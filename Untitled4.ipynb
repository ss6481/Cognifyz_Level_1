```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Dataset.csv')

# Display basic information about the DataFrame
print("DataFrame Info:")
print(df.info())

# first few rows
print(df.head())

cuisine_column = 'Cuisines'
city_column = 'City'

# Clean the data if necessary (e.g., handling missing values)
df[cuisine_column].fillna('Unknown', inplace=True)
df[city_column].fillna('Unknown', inplace=True)

# Aggregate data to find the top cuisines
top_cuisines = df[cuisine_column].value_counts().head(10)

# Aggregate data to find the top cities
top_cities = df[city_column].value_counts().head(10)

print("Top 10 Cuisines:")
print(top_cuisines)

print("Top 10 Cities:")
print(top_cities)

# Plotting the top cuisines
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values)
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()

# Plotting the top cities
plt.figure(figsize=(12, 6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities with Highest Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

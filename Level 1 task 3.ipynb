```python
import pandas as pd
df = pd.read_csv('Dataset.csv')
```

```python
#Visualize the locations of restaurants on a map using latitude and longitude information.

from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
```

```python
#using Shapety, Create Point geometry from Latitude and Longitude
gdf = gpd.GeoDataFrame (df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

#Create a base map of the world using Geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#Create a map that fits the screen and plots the restaurant Locations
ax = world.plot(figsize=(10, 6))
gdf.plot(ax=ax, column="continent", color = 'Orange', legend=True, figsize=(18, 15), marker="o", markersize=18)

import matplotlib.pyplot as plt

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of restaurants across different cities or countries

plt.figure(figsize=(9, 6))

#Due to large number of cities present,a set of top 5 are taken

sns.countplot(y = df['City'], order=df.City.value_counts().head(5).index, palette='Oranges')

plt.xlabel('Number of Restaurants')

plt.ylabel('Name of Cities')

plt.title('Distribution of Restaurants Across Cities')

plt.show()
```

```python
#Checking correlation between the restaurant's Location and its rating

plt.figure(figsize=(9, 5))

#Calculating the correlation between rating, longitude and latitude

correlation_matrix = df[['Aggregate rating', 'Latitude', 'Longitude']].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='pink', fmt=".2f")

plt.title("Correlation Between Restaurant's location and Rating")

plt.show()
```
